```python
import random as rd
import numpy as np
import math
import pylab as plt
```

# Question 1


```python
s = 5
np.random.seed(s)
N = 64
t = 10
dt = 1/64
ans = []
st = 0
time = np.arange(0,t,dt)
for i in range(10):
    for j in range(N):
        d = np.random.normal(0,1) 
        st += d 
        ans.append(st/8)
    
plt.plot(time,ans)
plt.grid()
```

# Question 2

```python
mu = 0.15
s = np.zeros(time.shape)
sigma = 0.4
S0 = 55.25

for i in range(len(time)):
    s[i] = S0*np.exp(   (mu - 0.5*sigma*sigma)*time[i]   + ans[i]*sigma)
    
plt.plot(time[:127],s[:127])
plt.grid()
```
